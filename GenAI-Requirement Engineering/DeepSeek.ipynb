```python
import os
from langchain_groq import ChatGroq
from langchain.prompts import ChatPromptTemplate


llm = ChatGroq(
    groq_api_key=os.environ["GROQ_API_KEY"],
    model_name='deepseek-r1-distill-qwen-32b'
)

prompt = ChatPromptTemplate.from_template("""
Introduction:
Act as a requirement analyst responsible for identifying and classifying stakeholders and end users for a Metro Ticket Distributor System.

Problem Statement:
The Metro Ticket Distributor System is being developed to provide an efficient and user-friendly experience for passengers purchasing tickets at metro stations. The system should support multiple ticketing options such as **single-trip, round-trip, and multi-trip tickets**, as well as **metro passes and time cards (daily, weekly, and monthly)**.
The system should handle various payment methods (cash, credit/debit cards, smartcards) and ensure fraud prevention by verifying currency authenticity.
It should also provide error handling mechanisms for:
- Aborted transactions
- Insufficient funds
- Change return and refund management
Additionally, the machine should operate efficiently under high traffic conditions, especially during peak hours on weekdays and weekends.

Task:
- Identify and categorize all relevant stakeholders who interact with or are impacted by the system.
- Differentiate between primary end users (direct users) and secondary stakeholders (indirect users).

Contextual Information:
- The Ticket Distributor System is designed to streamline ticket purchases and reduce queue times.
- It supports **cash, card, and metro pass transactions**.
- It must incorporate **security features, fraud prevention mechanisms, and accessibility options**.
- Stakeholders include **passengers, metro authorities, station staff, and maintenance teams**.

Instructions:
1. Classify stakeholders into primary (direct users) and secondary (indirect users).
2. For each stakeholder, describe:
   - Their role in using or managing the system.
   - Their expectations and concerns** regarding system functionality.
3. Identify potential conflicts** between stakeholders (e.g., security concerns vs. usability).
4. If applicable, consider **external regulatory authorities** involved in system governance.

Closing:
Once the stakeholders and end users have been identified, we will move to the next step: applying elicitation techniques to gather detailed requirements, ensuring a comprehensive understanding of their needs.
""")


prompt_template = prompt.format()
example = f"""
AI :
{prompt_template}

HUMAN :
Give me a list of "stakeholders" and "endUsers" separately.

AI :
"""


stakeholders_response = llm.invoke(example)
print(stakeholders_response.content)
```

```python
prompt = ChatPromptTemplate.from_template(f"""
**Introduction:**
Act as a **requirements analyst** responsible for selecting the most appropriate **elicitation techniques** for gathering requirements from the identified **stakeholders and end users**.

**Identified Stakeholders:**
{stakeholders_response.content}

**Task:**
1. Based on the given **stakeholders list**, suggest appropriate **elicitation techniques** for each stakeholder group.
2. Justify why the chosen technique is the best fit for gathering requirements from that stakeholder.
3. Provide a structured table format with:
   - **Stakeholder Name**
   - **Elicitation Techniques (e.g., interviews, surveys, observations, prototyping, focus groups, workshops)**
   - **Justification for Choosing the Technique**

**Contextual Information:**
- Some stakeholders (e.g., passengers) may require large-scale feedback methods like surveys, while others (e.g., metro staff) might be better suited for interviews.
- Government bodies and authorities may prefer formal documentation and reports.
- Metro authorities and payment service providers may need structured workshops or brainstorming sessions for technical and financial insights.
- Security and maintenance teams may require **observational** techniques to capture real-world challenges.

**Instructions:**
1. Assign at least **one elicitation technique per stakeholder**.
2. Justify why that technique is **suitable** based on their role, responsibilities, and influence over the system.

**Closing:**
Once the elicitation techniques are selected, the next step is to **apply them in real-world scenarios** to gather refined requirements for system design.
""")

prompt_template = prompt.format()
example = f"""
AI :
{prompt_template}

HUMAN :
Generate a structured list of elicitation techniques for each stakeholder with justifications.

AI :
"""


elicitation_techniques_response = llm.invoke(example)
print(elicitation_techniques_response.content)
```

```python
prompt = ChatPromptTemplate.from_template(f"""
**Introduction:**
Act as a requirements analyst responsible for explaining the effectiveness of the **elicitation techniques** chosen for different stakeholders in the NewYork TicketDistributor System.

**Elicitation Techniques Mapping:**
{elicitation_techniques_response.content}


**Task:**
For each stakeholder, provide:
1. A **detailed justification** of why the selected elicitation technique is the most suitable.
2. How the technique **helps in gathering system requirements effectively**.
3. If applicable, suggest any **alternative techniques** that could be used.

**Contextual Information:**
- **Surveys** are useful for large user bases (e.g., metro passengers) to gather **generalized insights**.
- **Interviews and Observations** help understand **real-world system interactions** (e.g., metro staff, maintenance teams).
- **Workshops and Brainstorming** sessions allow **collaborative discussions** for technical stakeholders (e.g., system administrators, payment providers).
- **Document Analysis** ensures compliance and aligns system features with **legal and regulatory requirements** (e.g., metro authorities).

**Instructions:**
1. Justify the selected technique for **each stakeholder** by explaining **why it is effective**.
2. Explain **how** the technique **extracts meaningful requirements** from that stakeholder.

**Closing:**
By applying the most appropriate elicitation techniques, we ensure that **all stakeholder concerns are captured accurately**. The next step is to **refine user stories and validate gathered requirements**.
""")

prompt_template = prompt.format()
example = f"""
AI :
{prompt_template}

HUMAN :
Provide a detailed justification of why each elicitation technique is effective and how it helps in gathering requirements.

AI :
"""

# Get the AI response
response = llm.invoke(example)
print(response.content)
```

```python
prompt = ChatPromptTemplate.from_template(f"""
**Introduction:**
Act as a **requirements analyst** responsible for defining the **functionalities of the Ticket Distributor System** in the form of **detailed user stories**.

**Stakeholders & Users:**
{stakeholders_response.content}

**Task:**
For each type of user, generate **at least 15 user stories**, ensuring that each user story contains a **detailed breakdown** of success and failure cases in the following format:

---

### **User Story Format:**

**Front of the Card:**
_As a [user role], I want to [goal] so that [benefit]._

**Back of the Card:**
1. **Success Scenarios:**
   - List all possible success cases where the feature works as expected.

2. **Failure Scenarios (with Display Messages):**
   - Include multiple possible failures such as system errors, incorrect user input, hardware malfunctions, and security-related constraints.
   - **Each failure scenario must be fully listed again instead of referring to previous examples.**
   - Each failure scenario must include the **display message shown to the user** for better UX clarity.

---

### **Example User Story:**

#### **Front of the Card:**
_As a passenger, I want to request assistance from a station attendant if I encounter difficulties so that I can receive help quickly._

#### **Back of the Card:**
1. **Success Scenarios:**
   (a) The assistance request is successfully sent, and a confirmation message is displayed:
      **"Your request has been sent. A station attendant will assist you shortly."**
   (b) The station attendant receives the request instantly on their device.
   (c) The request is assigned to an available attendant, and the estimated response time is displayed:
      **"An attendant will reach you in approximately 5 minutes."**
   (d) If no attendants are available, an automated response provides guidance or alternative contact options:
      **"No attendants are available right now. Please visit the help desk or call customer support."**

2. **Failure Scenarios (with Display Messages):**

   (a) **Network Issues:**
      - Poor internet connection prevents the request from being sent.
        **Display Message:** *"Network error! Please check your connection and try again."*
      - Request is delayed due to server congestion.
        **Display Message:** *"System is currently busy. Your request is in the queue. Please wait..."*

   (b) **System Errors:**
      - Request form does not load due to a backend failure.
        **Display Message:** *"Service unavailable! Please try again later."*
      - Attendant's system fails to acknowledge the request.
        **Display Message:** *"Request failed to reach the station attendant. Please try again."*

   (c) **User Errors:**
      - Passenger enters incorrect details, causing a delay.
        **Display Message:** *"Invalid information entered. Please review and submit again."*
      - Passenger mistakenly cancels the request before confirmation.
        **Display Message:** *"Request canceled! If you still need assistance, please submit a new request."*

   (d) **Security Concerns:**
      - A fake assistance request is detected and blocked.
        **Display Message:** *"Suspicious request detected. Assistance request denied."*
      - Unauthorized system access attempt triggers a security alert.
        **Display Message:** *"Unauthorized access attempt! This incident has been reported."*

---

### **Instructions:**
1. **Generate at least 15 user stories** for different users (e.g., Passenger, Metro Staff, Administrator, Maintenance Team).
2. **Ensure detailed success and failure cases** to comprehensively cover system functionalities.
3. **Each failure scenario must be listed explicitly again** with an appropriate error message displayed to the user.
4. **Do not reuse previous failure lists—each user story must have its own detailed breakdown.**

**Closing:**
By defining these functionalities in the form of structured **user stories**, we ensure a **clear understanding of system requirements**. The next step is to **validate and refine these stories before development**.
""")

prompt_template = prompt.format()
example = f"""
AI :
{prompt_template}

HUMAN :
Generate a **list of 20 user stories** for the Ticket Distributor System, ensuring **detailed success and failure scenarios** with display messages for each failure case.

AI :
"""

# Get the AI response
user_stories = llm.invoke(example)
print(user_stories.content)
```

```python
prompt = ChatPromptTemplate.from_template(f"""
**Introduction:**
Act as an **Agile Coach & Requirements Analyst**. Your task is to **validate the following user stories for the Ticket Distributor System** using the **INVEST framework** to ensure they align with agile best practices.

**Generated User Stories:**
{user_stories.content}

**Task:**
For each user story, apply the **INVEST framework validation** and provide a **detailed breakdown** of strengths and weaknesses using the following format:

---

### **INVEST Validation Format:**

#### **Front of the Card:**
_As a [user role], I want to [goal] so that [benefit]._

#### **Back of the Card:**
1. **Validation Results:**
   - **Independent:** (Pass/Fail) + Justification
   -  **Negotiable:** (Pass/Fail) + Justification
   - **Valuable:** (Pass/Fail) + Justification
   - **Estimable:** (Pass/Fail) + Justification
   -  **Small:** (Pass/Fail) + Justification
   - **Testable:** (Pass/Fail) + Justification

2. **Suggested Improvements (if needed):**
   - Provide specific recommendations if a criterion fails, ensuring the user story aligns with INVEST principles.

---

### **Instructions:**
1. **Validate all user stories provided above** based on the INVEST framework.
2. **List pass/fail for each criterion** with justifications.
3. **Suggest improvements** to align with INVEST principles.
4. **Validate all 20 user stories**.

**Closing:**
By validating these user stories against the **INVEST framework**, we ensure they are **clear, well-structured, and ready for agile development**.
""")

prompt_template = prompt.format()
example = f"""
AI :
{prompt_template}

HUMAN :
Validate the following all **user stories** for the Ticket Distributor System using the **INVEST framework**, and provide detailed feedback.

AI :
"""

# Get the AI response
invest_validations = llm.invoke(example)
print(invest_validations.content)
```

```python
validated_stories = []

# Split each user story validation using "---" as a delimiter
for story in invest_validations.content.split("---"):
    # Check if the user story **fully passes** the INVEST criteria
    if all(criteria in story for criteria in [
        "**Independent:** Pass",
        "**Negotiable:** Pass",
        "**Valuable:** Pass",
        "**Estimable:** Pass",
        "**Small:** Pass",
        "**Testable:** Pass"
    ]):
        validated_stories.append(story.strip())

# Store the validated user stories in a structured format
final_validated_output = "\n\n---\n\n".join(validated_stories)

# Print the validated user stories
print("**Validated User Stories:**\n")
print(final_validated_output)
```

```python
prompt = ChatPromptTemplate.from_template(f"""**Introduction:**
Act as an **Agile Coach & Requirements Analyst**. Your task is to **prioritize the validated user stories** for the **Ticket Distributor System** using the **MoSCoW method**, considering both **functional and non-functional aspects**.

---

### **Validated User Stories:**
{final_validated_output}

---

### **Task:**
1. **Analyze each validated user story** and prioritize it using the **MoSCoW method**:
   - **Must-have (M):** Critical for system functionality and must be implemented.
   - **Should-have (S):** Important but not immediately essential; can be deferred.
   - **Could-have (C):** Enhances usability or efficiency; nice-to-have but optional.
   - **Won’t-have (W):** Not needed in the current phase/version.

2. **Consider the following prioritization factors** while assigning MoSCoW categories:
   - **Business Value** (Impact on stakeholders and end users)
   - **Urgency** (How soon the feature is needed)
   - **Feasibility** (Technical complexity and implementation difficulty)
   - **Security & Compliance** (Risk, data protection, and regulatory impact)
   - **Performance & Usability** (Effect on system speed and user experience)
   - **Dependencies** (Enables or relies on other features)

3. **Output the prioritized user stories in the following format:**

---

### **MoSCoW Prioritization Format:**

#### **User Story:**
_As a [user role], I want to [goal] so that [benefit]._

#### **MoSCoW Classification:**
- **Priority:** Must-have (M) / Should-have (S) / Could-have (C) / Won’t-have (W)
- **Justification:** Explain why this priority was assigned based on the above factors.

---

### **Example Prioritization:**

#### **User Story:**
_As a passenger, I want to log in with my credentials so that I can access my account securely._

#### **MoSCoW Classification:**
- **Priority:** Must-have (M)
- **Justification:**
  - **Business Value:** Essential for authentication and security (5/5).
  - **Urgency:** Required before any other feature (5/5).
  - **Feasibility:** Well-established implementation methods (5/5).
  - **Security & Compliance:** High (5/5) as it involves user credentials.
  - **Performance & Usability:** Low impact on performance (4/5).
  - **Dependencies:** Needed for all account-related features (5/5).

---

### **Instructions:**
1. **Apply MoSCoW prioritization** to all validated user stories.
2. **Justify each priority level** based on business value, urgency, feasibility, security, performance, and dependencies.
3. **Ensure security, compliance, and performance considerations are reflected.**
4. **Structure the output clearly, as shown in the example.**

---

**Closing:**
By prioritizing user stories with the **MoSCoW method**, we ensure that critical features are delivered first while balancing feasibility, security, and business impact.

""")

prompt_template = prompt.format()
example = f"""
AI :
{prompt_template}

HUMAN :
Using the **MoSCoW method**, prioritize the following validated user stories for the **Ticket Distributor System**, considering both functional and non-functional aspects such as security, performance, feasibility, urgency, and business value.

AI :
"""

# Get the AI response
prioritize = llm.invoke(example)
print(prioritize.content)
```

```python
prompt = ChatPromptTemplate.from_template(f"""

### **Validated User Stories:**
{final_validated_output}



### **Task:**
1. **Group related user stories into EPICs** (broad functionalities) and identify at least **three conflicting EPICs**.
2. **Describe the nature of the conflict** between requirements within each EPIC.
   - **Types of conflicts:**
     - Functional conflicts (e.g., conflicting feature priorities)
     - Non-functional conflicts (e.g., security vs. usability)
     - Resource conflicts (e.g., performance vs. cost)
     - Stakeholder conflicts (e.g., different user needs)
3. **Suggest resolution strategies** for each conflict, ensuring feasibility and alignment with business objectives.
4. Format the output as follows:

---

### **Example Output:**

#### **EPIC 1: Secure User Authentication**
- **Conflicting User Stories:**
  - _As a passenger, I want to log in with biometric authentication so that I can access my account quickly._
  - _As an admin, I want strict two-factor authentication so that unauthorized users cannot access the system._
- **Conflict Type:** Usability vs. Security
- **Conflict Analysis:**
  - Biometric authentication improves usability but may not meet regulatory compliance for strict security.
  - Two-factor authentication enhances security but might frustrate frequent passengers due to additional steps.
- **Resolution Strategy:**
  - Implement **adaptive authentication** where high-risk actions require 2FA, while low-risk logins allow biometrics.

---

#### **EPIC 2: Real-Time Ticket Processing**
- **Conflicting User Stories:**
  - _As a passenger, I want real-time ticket processing so that I don’t have to wait in queues._
  - _As a system architect, I want to optimize server load to prevent crashes during peak hours._
- **Conflict Type:** Performance vs. Scalability
- **Conflict Analysis:**
  - Instant ticket processing puts a heavy load on servers, potentially causing system crashes during rush hours.
  - Delayed processing reduces load but negatively impacts user experience.
- **Resolution Strategy:**
  - Implement **queue-based ticketing with caching** to balance performance and system stability.

---

#### **EPIC 3: Refund & Cancellation Policies**
- **Conflicting User Stories:**
  - _As a passenger, I want instant refunds for canceled tickets so that I can reuse my money immediately._
  - _As a finance manager, I want refund approvals to be manually verified to prevent fraud._
- **Conflict Type:** Speed vs. Fraud Prevention
- **Conflict Analysis:**
  - Instant refunds improve customer satisfaction but may lead to financial fraud or abuse.
  - Manual verification ensures security but slows down the refund process.
- **Resolution Strategy:**
  - Implement **automated fraud detection** for low-risk transactions while keeping manual verification for high-risk refunds.

---

### **Instructions:**
1. Identify **three EPICs** where requirement conflicts occur.
2. Describe the **conflict type** and **analysis** of the problem.
3. Propose **resolution strategies** that balance competing needs.

**Closing:**
By resolving these requirement conflicts, we ensure a **balanced and effective system design** that meets both user needs and technical constraints.

""")

prompt_template = prompt.format()
example = f"""
AI :
{prompt_template}

HUMAN :
Analyze the **validated user stories** and identify **three EPICs** (or collections of related user stories) where conflicts between requirements exist in the **Ticket Distributor System**.

AI :
"""

# Get the AI response
epics = llm.invoke(example)
print(epics.content)
```
